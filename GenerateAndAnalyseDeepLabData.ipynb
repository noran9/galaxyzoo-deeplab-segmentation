```python
df = pd.read_csv('/home/noranikoloska/Documents/MasinskaVizija/training_solutions_rev1.csv')
df.head()

# Calculate Max Class

df['MaxClass'] = df.loc[:, df.columns != 'GalaxyID'].idxmax(axis=1)
df['MaxClass'] = df['MaxClass'].str.replace('Class', '')
df.head()
```

```python
print(df.GalaxyID.to_string(index=False))
```

```python
trainingImagesPath = '/home/noranikoloska/Documents/MasinskaVizija/training'

segmentedImagesPath = '/home/noranikoloska/Documents/MasinskaVizija/Tensorflow/models/research/deeplab/datasets/GalaxyZoo/SegmentationClass/'
jpegImagesPath = '/home/noranikoloska/Documents/MasinskaVizija/Tensorflow/models/research/deeplab/datasets/GalaxyZoo/JPEGImages/'

label_colours = [(0,0,0)
                # 0 = background
                ,(128,0,0),(0,128,0),(128,128,0),(0,0,128),(128,0,128)]
                # 1 = Class1.1, 2 = Class1.2, 3 = Class1.3, 4 = Class6.1, 5 = Class6.2

for fileName in listdir(trainingImagesPath):
    
    # Calculate class and color
    imageClass = df.loc[df['GalaxyID'] == int(fileName.replace('.jpg', ''))]['MaxClass'].values[0]
    if(imageClass == '1.1'):
        color = label_colours[1]
    elif(imageClass == '1.2'):
        color = label_colours[2]
    elif(imageClass == '1.3'):
        color = label_colours[3]
    elif(imageClass == '6.1'):
        color = label_colours[4]
    else:
        color = label_colours[5]
    
    # Load image
    img = cv2.imread(trainingImagesPath + "/" + fileName)
    segmentedImage = img.copy()
    
    # Create segmented image
    grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    retval, threshold = cv2.threshold(grayImage, 50, 255, cv2.THRESH_BINARY)
    segmentedImage[threshold == 255] = color
    segmentedImage[threshold != 255] = label_colours[0]

    # For visualisation purposes
    # plt.imshow(segmentedImage, cmap="gray"), plt.show()
    
    # Save images in designated folders
    # Note: the image file name must remain the same
    cv2.imwrite(jpegImagesPath + fileName, img)
    cv2.imwrite(segmentedImagesPath + fileName, cv2.cvtColor(segmentedImage, cv2.COLOR_RGB2BGR))
```

```python
# Dice coefficient evaluation

imagesPath = '/home/noranikoloska/Documents/MasinskaVizija/vis11/segmentation_results'

score = 0
count = 0

for fileName in listdir(imagesPath):
    if ('image' in fileName):
        
        #Compute ground truth:
        img = cv2.imread(imagesPath + '/' + fileName)        

        grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        retval, threshold = cv2.threshold(grayImage, 50, 255, cv2.THRESH_BINARY)
        img[threshold == 255] = [128,0,0]
        img[threshold != 255] = [0, 0, 0]
                
        #Extract segmented image path
        predictedImage = fileName[0:6] + '_prediction.png'
        predicted = cv2.imread(imagesPath + '/' + predictedImage)
        predicted = cv2.cvtColor(predicted, cv2.COLOR_BGR2RGB)
        
        #Compute dice coefficient
        img1 = np.asarray(img).astype(np.bool)
        img2 = np.asarray(predicted).astype(np.bool)
                        
        intersection = np.logical_and(img1, img2)
        score += (2. * intersection.sum() / (img1.sum() + img2.sum()))
        count += 1
        
print(score / count)
```
